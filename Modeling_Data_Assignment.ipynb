# Question 1: What is Data Loading in Power BI and why is it considered the first step of analysis?

## Answer:

Data Loading in Power BI is the process of importing data from different data sources into Power BI for analysis and reporting.

The data can be loaded from:
- CSV files
- Excel files
- SQL databases
- Cloud platforms
- Web sources

### Why it is considered the first step of analysis:

1. Analysis cannot begin without data.

2. It helps Power BI connect with raw business data.

3. Reports and dashboards are created only after loading the dataset.

4. Proper loading ensures accurate analysis and visualization.

### Conclusion:

Data loading is the foundation of Power BI because all business insights depend on the imported data.


# Question 2: Explain the difference between “Load” and “Transform Data” in Power BI.

## Answer:

Both “Load” and “Transform Data” are options available while importing data into Power BI.

| Basis | Load | Transform Data |
|-------|------|----------------|
| Meaning | Imports data directly into Power BI | Opens Power Query Editor for cleaning |
| Purpose | Quick loading of clean data | Cleaning and transforming data |
| Modification | No changes are made | Data can be modified |
| Usage | Used for already cleaned data | Used when preprocessing is needed |

### Example:
- Load → Import clean CSV file directly.
- Transform Data → Remove duplicates, handle null values, and change formats.

### Conclusion:

“Load” imports data directly, while “Transform Data” is used to clean and prepare data before analysis.


# Question 3: What is a Fact Table and a Dimension Table? Give examples from the dataset.

## Answer:

### Fact Table:

A Fact Table contains measurable numerical values used for analysis.

### Examples from Dataset:
- Confirmed Cases
- Deaths
- Recovered Cases

These columns contain business metrics and calculations.


### Dimension Table:

A Dimension Table contains descriptive information used for filtering and grouping data.

### Examples from Dataset:
- Country
- State
- Date
- Vaccination_Status

These columns provide context to the numerical data.


### Difference between Fact Table and Dimension Table:

| Fact Table | Dimension Table |
|------------|----------------|
| Contains numerical values | Contains descriptive data |
| Used for calculations | Used for filtering |
| Example: Confirmed Cases | Example: Country |

### Conclusion:

Fact tables store measurable data, while dimension tables describe and organize the data.


# Question 4: Why is Star Schema preferred over Snowflake Schema in Power BI?

## Answer:

Star Schema is preferred in Power BI because it is simple, fast, and improves dashboard performance.

### Reasons:

1. Easy to Understand  
   Star schema has a simple structure.

2. Faster Performance  
   Fewer table joins improve report speed.

3. Better Dashboard Efficiency  
   Power BI works efficiently with star schema models.

4. Easy Relationship Management  
   Relationships between tables are easier to maintain.


### Difference:

| Star Schema | Snowflake Schema |
|-------------|-----------------|
| Simple structure | Complex structure |
| Faster queries | More joins required |
| Easy to maintain | Harder to manage |

### Conclusion:

Star schema improves performance and simplifies data modelling in Power BI.


# Question 5: Identify and remove duplicate records based on Date, Country, and State.

## Answer:

### Steps to Remove Duplicate Records:

1. Open Power Query Editor.

2. Select the following columns:
- Date
- Country
- State

3. Go to:
### Home Tab → Remove Rows → Remove Duplicates

4. Power BI will automatically remove duplicate records.

### Purpose:
Removing duplicates improves data accuracy and prevents repeated calculations.

### Conclusion:

Duplicate removal ensures correct reporting and reliable analysis.


# Question 6: Identify and replace null values in Vaccination_Status.

## Answer:

### Steps to Replace Null Values:

1. Open Power Query Editor.

2. Select the Vaccination_Status column.

3. Identify rows containing null values.

4. Right-click the column.

5. Select:
### Replace Values

6. Replace:
- Null → “Not Vaccinated”

### Why this is important:
- Prevents missing data issues
- Improves report accuracy
- Makes analysis easier

### Conclusion:

Replacing null values improves data quality and reporting reliability.
